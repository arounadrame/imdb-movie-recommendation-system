```python
import pandas as pd
```

```python
df=pd.read_csv("movies.csv")
```

```python
df_clean = df[
    [
        "Series_Title",
        "Overview",
        "Genre",
        "Director",
        "Star1",
        "Star2",
        "Star3",
        "Star4",
        "IMDB_Rating",
        "Poster_Link"
    ]
]
```

```python
df_clean.isna().sum()
```

```python
df_clean["combined_features"] = (
    df["Overview"] + " " +
    df["Genre"] * 2 + " " +
    df["Director"] * 3  + " " +
    df["Star1"] * 2 + " " +
    df["Star2"] * 2 + " " +
    df["Star3"] + " " +
    df["Star4"]
)
df_clean[["Series_Title", "combined_features"]].head()
```

```python
df_clean.to_csv("movies_cleaned.csv",index=False)
```
