```python
import pandas as pd
```

```python
df=pd.read_csv("IMDB.csv")
df2=pd.read_csv("imdb_top_1000.csv")
```

```python
df = df.merge(
    df2[["Series_Title", "Poster_Link"]],
    on="Series_Title",
    how="left"
)
```

```python
df.head()
```

```python
df.to_csv("movies.csv",index=False)
```

```python
df_clean = df[
    [
        "Series_Title",
        "Overview",
        "Genre",
        "Director",
        "Star1",
        "Star2",
        "Star3",
        "Star4",
        "IMDB_Rating",
        "Poster_Link"
    ]
]
```

```python
df_clean.isna().sum()
```

```python
df_clean["combined_features"] = (
    df["Overview"] + " " +
    df["Genre"] + " " +
    df["Director"] + " " +
    df["Star1"] + " " +
    df["Star2"] + " " +
    df["Star3"] + " " +
    df["Star4"]
)
df_clean[["Series_Title", "combined_features"]].head()
```

```python
df_clean.to_csv("movies_cleaned.csv",index=False)
```
